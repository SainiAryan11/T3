```python
data = {"name":[],"price":[],"reviews":[]}
for i in box:
    data["name"].append(i.find('a',class_='title').text.strip())
    data['price'].append(i.find('span').text.strip())
    data['reviews'].append(i.find('p',class_='review-count').find('span').text.strip())
import pandas as pd
df = pd.DataFrame(data)
df
```
